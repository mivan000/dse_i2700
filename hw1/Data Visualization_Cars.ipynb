```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
raw_data = 'http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
cols_link = 'http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names'
```

```python
columns = ['symboling',
'normalized-losses',
'make',
'fuel-type',
'aspiration',
'num-of-doors',
'body-style',
'drive-wheels',
'engine-location',
'wheel-base',
'length',
'width',
'height',
'curb-weight',
'engine-type',
'num-of-cylinders',
'engine-size',
'fuel-system',
'bore',
'stroke',
'compression-ratio',
'horsepower',
'peak-rpm',
'city-mpg',
'highway-mpg',
'price']
```

```python
cars = pd.read_csv(raw_data, na_values='?', names = columns)
```

```python
mapping = {'four': 4, 'six': 6, 'five': 5, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8}
```

```python
cars['num-of-cylinders-numerical'] = cars['num-of-cylinders'].replace(mapping)
```

```python
cars[['num-of-cylinders','num-of-cylinders-numerical']]
```

```python
cars.columns
```

```python
cars_vega = cars[['make', 'fuel-type', 'body-style', 'drive-wheels','curb-weight', 'engine-type',
       'num-of-cylinders', 'num-of-cylinders-numerical', 'engine-size', 'horsepower','city-mpg', 'highway-mpg', 'price']]
```

```python
# columns wiht null values
cars_null = cars_vega[cars_vega.isnull().values]
cars_null
```

```python
# flagging rows with nulls
cars_vega_w_null = cars_vega[~cars_vega.isin(cars_null)]
cars_vega_w_null
```

```python
cars_vega_no_null = cars_vega_w_null.dropna()
cars_vega_no_null
```

```python
cars_vega_no_null.isna().sum()
```

```python
cars_vega_no_null.to_csv('cars_parallel.csv', index = False)
```
